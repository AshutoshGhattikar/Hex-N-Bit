Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
# importing needed libraries
import matplotlib.pyplot as plt
import numpy as np
```

### Sample Data

```python
interval_x=(np.linspace(0,20,50))
test_data_y=interval_x              # y=x
squared_test_data_y=interval_x**2   # y=(x*x)
cubic_test_data_y=interval_x**3     # y=(x*x*x)
#interval_x.shape
```

## Plotting Data

```python
plt.plot(interval_x,test_data_y)   # plot x and y... test_data_y on y axis and interval_x on x axis
#plt.show()
```

### Changing Colour of Plot

Check supported colours on following link<br>
https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
# change colour of plot
plt.plot(interval_x,test_data_y,'g')  # here g is to plot using green colour
#plt.show()
```

### Changing Style of Plot

```python
# Use Dark Background
plt.style.use('dark_background')
#print(plt.style.available)
# change colour of plot
plt.plot(interval_x,test_data_y,'w')  # here g is to plot using green colour
#plt.show()
```

### Labeling Axes and Plot

```python
plt.style.use('default')
plt.plot(interval_x,test_data_y,'y')  # here y is to plot using yellow colour
plt.xlabel("X Axis")                  # labeling X Axis
plt.ylabel("Y Axis")                  # labeling Y Axis
plt.title("Linear Graph")             # Naming the graph
#plt.show()
```

### Change Size of Figure

```python
plt.rcParams["figure.figsize"] = [6,4]
```

### Line Width and Line Styles

Refer link below to get an idea about all types of supported linestyles <br>
https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html

```python
plt.plot(interval_x,test_data_y,'y',linewidth=5)  # here y is to plot using yellow colour
#plt.show()
```

```python
plt.plot(interval_x,test_data_y,'y',linewidth=4,linestyle='dotted')  # dotted line
#plt.show()
```

```python
plt.plot(interval_x,test_data_y,'y',linewidth=3,linestyle='dashed')  # dashed line
#plt.show()
```

```python
plt.plot(interval_x,test_data_y,'y',linewidth=2,linestyle='dashdot')  # dashdot lime
#plt.show()
```

### Marker Types and Marker Size

Refer link below to get an idea about all types of supported markers <br>
https://matplotlib.org/3.2.1/api/markers_api.html

```python
plt.plot(interval_x,test_data_y,'g*',markersize=7)  # star shaped marked, # 'g*' means green stars
#plt.show()
```

```python
plt.plot(interval_x,test_data_y,'go',markersize=2)# circle shaped marked, # g in 'go' means green and o in "go" means circles
#plt.show()
```

### Multiple Plots

```python
# plt.subplot(nrows,ncolumns,plot_index)
plt.subplot(2,1,1)                               # subplot number 1 on a plot of 2 rows and 1 column
plt.plot(interval_x,squared_test_data_y, 'g-.')  # dash dot line style marker in green
plt.xlabel("X")
plt.ylabel("Y=X Square")
plt.title("Square")

plt.subplots_adjust(hspace=1)  # to insert some space between both sub plots

plt.subplot(2,1,2)                               # subplot number 2 on a plot of 2 rows and 1 column
plt.plot(interval_x,cubic_test_data_y,'r--')     # dashed line style marker in red
plt.xlabel("X")
plt.ylabel("$Y^3$")   # another way of writing  (TeX style markup)
plt.title("Cubic")
#plt.show()
```

### Adding Legend

https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html

```python
# plt.subplot(nrows,ncolumns,plot_index)
plt.subplot(1,2,1)                                             # subplot number 1 on a plot of 2 rows and 1 column
plt.plot(interval_x,squared_test_data_y,'g-.',label='Square')  # Label will be added in the legend
plt.xlabel("X")
plt.ylabel("Y=X Square")
plt.title("Square")
plt.legend()                                                   # to display legend

plt.subplots_adjust(wspace=1)                                  # to insert some space between both sub plots

plt.subplot(1,2,2)                                             # subplot number 2 on a plot of 2 rows and 1 column
plt.plot(interval_x,cubic_test_data_y,'r--',label='Cubic')     # Label will be added in the legend
plt.xlabel("X")
plt.ylabel("Y=X Cube")
plt.title("Cubic")
plt.legend(loc='upper center')                                 # to display legend, location can also be changed
# the position of legend can also be changed by using numbers corresponding to the position strings
# visit the link in above heading markdown for details
#plt.show()
```

### Adding Text on Plot

Visit https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.text.html for more details

```python
plt.plot(interval_x,test_data_y,'y',linewidth=2,linestyle='dashdot')  # dashdot lime
plt.text(0,18,"Dummy Text",fontsize=20)    # plt.text(xpos,ypos,"Text",fontsize)
#plt.show()
```

### Example Showing Various Line Types and Marker Types

Reference: https://www.southampton.ac.uk/~feeg1001/notebooks/Matplotlib.html

```python

x = np.linspace(0, 5, 10)
y = x ** 2


fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# possible linestype options ‘-‘, ‘--’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='--', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='--', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='--', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue")
```
