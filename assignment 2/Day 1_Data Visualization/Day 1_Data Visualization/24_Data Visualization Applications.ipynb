```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

### Reading Data from "car_sales.csv"

```python
data=pd.read_csv("car_sales.csv")
pd.set_option('display.max_columns', None) # to display all columns
# pd.set_option('display.max_rows', None)  # uncomment to display all rows
data.head()  # looking at the first 5 rows
```

```python
# Plotting Horsepower vs. Mileage on Highway

fig,axes=plt.subplots()
axes.plot(data["Horsepower"],data["MPG.highway"],"r--")
axes.set_xlabel("Horsepower",fontsize=13)                               # labelling X Axis
axes.set_ylabel("Mileage on Highway",fontsize=13)                       # labelling Y Axis
axes.set_title("Horsepower vs. Mileage",fontsize=15)                    # Title for plot
```

```python
# Sorting so that line plot is proper when plotted. Otherwise they will make no sense, try on your own without sorting
# Scatter plot should run fine without sorting as one point is not connected to other
data_sorted=data.sort_values("Horsepower",ascending=True)
```

```python
data_sorted.head()
```

```python
# Plotting Horsepower vs. Mileage on Highway

fig,axes=plt.subplots()
axes.plot(data_sorted["Horsepower"],data_sorted["MPG.highway"],"r--")   # plotting
axes.set_xlabel("Horsepower",fontsize=13)                               # labelling X Axis
axes.set_ylabel("Mileage on Highway",fontsize=13)                       # labelling Y Axis
axes.set_title("Horsepower vs. Mileage",fontsize=15)                    # Title for plot
```

```python
# Plotting with Circle type marker, all other parameters same as above
# This case onwards, we are not using sorted data
fig,axes=plt.subplots()
axes.plot(data["Horsepower"],data["MPG.highway"],"ro")
axes.set_xlabel("Horsepower",fontsize=13)
axes.set_ylabel("Mileage on Highway",fontsize=13)
axes.set_title("Horsepower vs. Mileage",fontsize=15)
```

### Scatter Plot

```python
# Scatter plot is equivalent to plot with circle type markers (at least in this case)
fig,axes=plt.subplots()
axes.scatter(data["Horsepower"],data["MPG.highway"],c="g")
axes.set_xlabel("Horsepower",fontsize=13)
axes.set_ylabel("Mileage on Highway",fontsize=13)
axes.set_title("Horsepower vs. Mileage",fontsize=15)
```

### Bar Graph

```python
fig,axes=plt.subplots()
axes.bar(data["Passengers"],data["Passengers"].value_counts()[data["Passengers"]],color="yellow",edgecolor="red")
# in the above statement, we are plotting "Number of Passenges" on X and Number of entries for them on Y
# print data["Passengers"] and then data["Passengers"].value_counts()[data["Passengers"] to understand clearly
# Color and edgecolor are just optional, to show the capability.
axes.set_xlabel("Number of Passengers",fontsize=13)
axes.set_ylabel("Cars in Data",fontsize=13)
axes.set_title("Passenger Count Data",fontsize=15)
```

## Line Plot

```python
# To plot Horsepower vs. Mileage.. as we did above
new_plot=data_sorted.plot.line(x="Horsepower",y="MPG.highway")
# new_plot can be used to set other features of plot if needed
new_plot.set_title("Mileage vs. Horsepower")
```
