# Matplotlib Object Oriented API

In Object Oriented API we directly work with the objects, in order to have more control and customization of our plots.

In object-oriented programming we have objects on which we apply functions and actions. The has advantages when we have more than one figures to be created, or when a the figures contains more than one subplot.

Examples: https://matplotlib.org/gallery/index.html#examples-index<br>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

### Sample Data

```python
x=np.arange(0,20)  
y=x**2             # y=x^2
```

## Plotting Data

```python
fig = plt.figure()  # creates blank figure
# adding new set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9]) #[left, bottom, width, height] of the new axes. 
                                          #All quantities are in fractions of figure width and height
axes.plot(x,y)
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
axes.set_title('Square')
```

### Multiple Plots in one Figure, custom axes

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 0.5, 0.5])    # adding axes
axes2 = fig.add_axes([0.7, 0, 1.2, 1.2])  # adding axes

axes1.plot(x,y)
axes1.set_xlabel('X-Axis')
axes1.set_ylabel('Y-Axis')
axes1.set_title('Square') 


axes2.plot(x,y,"ro")        # red circles
axes2.set_xlabel('X-Axis')
axes2.set_ylabel('Y-Axis')
axes2.set_title('Square')
```

```python
# Axes withing Axes
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 0.9, 0.9]) 
axes2 = fig.add_axes([0.15, 0.5, 0.3, 0.3]) 

axes1.plot(x,y)
axes1.set_xlabel('X-Axis')
axes1.set_ylabel('Y-Axis')
axes1.set_title('Square')


axes2.plot(x,y,"ro")
axes2.set_xlabel('X-Axis')
axes2.set_ylabel('Y-Axis')
axes2.set_title('Square')
```

### Multi Plot using Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=2) # axes is a list object
axes[0][0].plot(x,y,"r")                   # axes has been indexed for subplot 1
axes[0][0].set_xlabel("X-Axis")
axes[0][0].set_ylabel("Y-Axis")

axes[0][1].plot(x,y,"g")                   # axes has been indexed for subplot 2
axes[0][1].set_xlabel("X-Axis")
axes[0][1].set_ylabel("Y-Axis")

axes[1][0].plot(x,y,"b")                   # axes has been indexed for subplot 3
axes[1][0].set_xlabel("X-Axis")
axes[1][0].set_ylabel("Y-Axis")
 
axes[1][1].plot(x,y,"y")                   # axes has been indexed for subplot 4
axes[1][1].set_xlabel("X-Axis")
axes[1][1].set_ylabel("Y-Axis") 
```

### Layout in order to fix overlap

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y,"r")
axes[0][0].set_xlabel("X-Axis")
axes[0][0].set_ylabel("Y-Axis")

axes[0][1].plot(x,y,"g")
axes[0][1].set_xlabel("X-Axis")
axes[0][1].set_ylabel("Y-Axis")

axes[1][0].plot(x,y,"b")
axes[1][0].set_xlabel("X-Axis")
axes[1][0].set_ylabel("Y-Axis")

axes[1][1].plot(x,y,"y")
axes[1][1].set_xlabel("X-Axis")
axes[1][1].set_ylabel("Y-Axis")

fig.tight_layout()                   # to fix overlaps
```

### Custom Layout

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y,"r")
axes[0][0].set_xlabel("X-Axis")
axes[0][0].set_ylabel("Y-Axis")

axes[0][1].plot(x,y,"g")
axes[0][1].set_xlabel("X-Axis")
axes[0][1].set_ylabel("Y-Axis")

axes[1][0].plot(x,y,"b")
axes[1][0].set_xlabel("X-Axis")
axes[1][0].set_ylabel("Y-Axis")

axes[1][1].plot(x,y,"y")
axes[1][1].set_xlabel("X-Axis")
axes[1][1].set_ylabel("Y-Axis")

fig.subplots_adjust(hspace=0.6,wspace=0.6)  # adds custom spacing between plots
# hspace is basically "height spacing" and wspace is "width spacing"
#fig.subplots_adjust(hspace=0.6)
#fig.subplots_adjust(wspace=0.6) 
```

### Custom Figure Size

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(7,5), dpi=150) # figsize and dpi parameters
axes[0][0].plot(x,y,"r")
axes[0][0].set_xlabel("X-Axis")
axes[0][0].set_ylabel("Y-Axis")
axes[0][0].set_title("Square")

axes[0][1].plot(x,y,"g")
axes[0][1].set_xlabel("X-Axis")
axes[0][1].set_ylabel("Y-Axis")
axes[0][1].set_title("Square")

axes[1][0].plot(x,y,"b")
axes[1][0].set_xlabel("X-Axis")
axes[1][0].set_ylabel("Y-Axis")
axes[1][0].set_title("Square")

axes[1][1].plot(x,y,"y")
axes[1][1].set_xlabel("X-Axis")
axes[1][1].set_ylabel("Y-Axis")
axes[1][1].set_title("Square")

fig.subplots_adjust(hspace=0.6)
fig.subplots_adjust(wspace=0.6)
```

##### Sample

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
```

### Legend and Legend Position

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r",label="Plot1")  # use label parameter, will be added in legend
axes.plot(y,x,"g",label="Plot2")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.legend()       # to show legend
#axes.legend(loc=0) # let matplotlib decide the optimal location
#axes.legend(loc=1) # upper right corner
#axes.legend(loc=2) # upper left corner
#axes.legend(loc=3) # lower left corner
#axes.legend(loc=4) # lower right corner
```

### Axis Limit

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r",label="Plot1")
axes.plot(y,x,"g",label="Plot2")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.set_ylim(0,500)  # change y axis limits
axes.set_xlim(0,200)  # change x axis limits 
axes.legend()  
```

### Padding for Axis Label & Tickers on axes

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r",label="Plot1")
axes.plot(y,x,"g",label="Plot2")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.xaxis.labelpad = 10  # change padding for x axis label
axes.yaxis.labelpad = 20  # change padding for y axis label

axes.set_xticks(np.arange(0,400,30)) # change tickers for x axis at interval of 30
axes.set_yticks(np.arange(0,460,30)) # change tickers for y axis at interval of 30
axes.legend()
```

### Saving Plot

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
fig.savefig("filename.png")   # used to save as an image file
#fig.savefig("filename.png",dpi=300)
```

### Enabling Grids

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.grid(True)   # enables the grid if parameter is true
#axes.grid(color='r', alpha=1, linestyle='dotted', linewidth=1)
```

### Spines

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.grid(True)

axes.spines['left'].set_color('Yellow')  # change color of specified spine
axes.spines['left'].set_linewidth(2)     # change width of specified spine
axes.spines['top'].set_color('Red')
axes.spines['top'].set_linewidth(2)
axes.spines['bottom'].set_color('Green')
axes.spines['bottom'].set_linewidth(2)
#axes.grid(color='r', alpha=1, linestyle='dotted', linewidth=1)
```

### Major Minor Grids and Ticks

```python
from matplotlib.ticker import MultipleLocator  # import multiplelocator
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.plot(y,x,"g")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.xaxis.set_major_locator(MultipleLocator(100))  # set interval of major grid line
axes.xaxis.set_minor_locator(MultipleLocator(20))   # set interval of minor grid line
axes.grid(True,which='major',color="pink")          # enable minor grids
axes.grid(True,which='minor',color="black")         # enable major grids
```

### Text

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,"r")
axes.set_xlabel("X-Axis")
axes.set_ylabel("Y-Axis")
axes.set_title("Square")
axes.text(9,170,"$y=x^2$",fontsize=20,color="red")  # insert text
```
