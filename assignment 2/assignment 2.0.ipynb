### Task 1
In a single plot, plot the following:-
1. Plot Sin wave from 0-360 degrees. (Degrees on X axis and value of sin on Y axis)
2. Plot Cos wave from 45-360 degrees. (Degrees on X axis and value of cos on Y axis)
3. Add legend to the plot
4. Add Axis Labels
5. Add Plot Title as "Trigonometric Plots"
6. Add gridlines to the plot

```python
import matplotlib.pyplot as plt
import numpy as np       #importing numpy as np
x = np.arange(0,361,1)   #arange function is used to get the values from 0 to 360
y = np.sin(x)            
a = np.arange(45,361,1)   #arange function is used to get the values from 45 to 360 
b = np.cos(a) 


                             
plt.plot(x,y,'b',label='Sin(x)')
plt.xlabel("Degrees ->") 
plt.ylabel("Sin")
plt.title("Trigonometric Plots")
plt.legend(loc='center') 
plt.grid(True) 
                            
```

```python

plt.plot(a,b,'y',label='Cos(x)')  
plt.xlabel("Degrees ->")
plt.ylabel("Cos")
plt.title("Trigonometric Plots")
plt.legend(loc='center') 
plt.grid(True) 
```
