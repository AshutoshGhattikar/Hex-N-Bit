### Task 3

Plot the Elbow Plot in  order to implement K-Means Clustering and write the value of selected K

```python
x = df.iloc[:, [3, 4]].values
```

```python
distortions = []
K = range(1,11)   
for k in K:
    kmeanModel = KMeans(n_clusters=k)  
    kmeanModel.fit(x)                   
    distortions.append(kmeanModel.inertia_)   
plt.plot(K, distortions, 'go-')
plt.xlabel('Value of K')
plt.ylabel('Distortion')
plt.title('Elbow Plot')
```

```python
km = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)
y_means
```

### Task 4

With selected K value, perform K-Means Clustering to cluster the customers

```python
km = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)
```

### Task 5





Add the predicted cluster labels to the main dataframe as a new column and display the complete dataframe

### Task 6

Create a scatter plot (Annual Income  vs Spending Score) and colour code the scatter plot as per the clustered customers<br>
**HINT :** df.plot.scatter

```python
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'green', label = 'J')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'A')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'cyan', label = 'V')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s = 100, c = 'magenta', label = 'A')
plt.scatter(x[y_means == 4, 0], x[y_means == 4, 1], s = 100, c = 'orange', label = 'S')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```
