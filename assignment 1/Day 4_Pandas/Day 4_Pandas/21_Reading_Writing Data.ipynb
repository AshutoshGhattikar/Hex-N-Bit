# Reading Writing Data from External Sources

It is possible to read and write data from and to various external sources, few of these sources can be:-
    - Text Files
    - CSV Files
    - HTML Pages
    - Excel Files
    - JSON Files
    - SQL

```python
import pandas as pd
import numpy as np
```

## CSV Files

### Reading

```python
# reading from csv file placed in project directory- File Name: car_sales.csv
car_data=pd.read_csv("car_sales.csv")
type(car_data)
```

```python
car_data
```

```python
# printing Airbags feature of BMW from the data
car_data[car_data["Manufacturer"] == "BMW"]["AirBags"]
```

### Writing

```python
# producing dummy data which will be written in file
dummy_data=pd.DataFrame(np.arange(15).reshape(5,3),columns="C1 C2 C3".split())
dummy_data
```

```python
# writing dummy data to Dummy_data.csv
#dummy_data.to_csv("Dummy_data.csv")
dummy_data.to_csv("Dummy_data.csv",index=False)
```

## Excel Files

## Reading

```python
# reading from excel SampleData.xlsx and sheet name SalesOrder
sample_data=pd.read_excel('SampleData.xlsx',sheet_name="SalesOrders")
type(sample_data)
```

```python
sample_data
```

### Writing Data

```python
# Writing Dummy Data to file Excel_Sample.xlsx on sheet name Sheet1
dummy_data.to_excel('Excel_Sample.xlsx',sheet_name='Sheet1',index=False)
```

```python
print(dummy_data)
```
